## Exercise - Inheritance 

In addition to your `student.py`, create a `staff.py` file which contains a `Staff` class. 

Create a `person.py` file, in which a `Person` class is the superclass of `Student` and `Staff`. 

Write your classes so that the `Student` and `Staff` inherit common attributes and methods from the `Person` class. 


````python
# Write your solution here or in a dedicated py file.
class Person:
    def __init__(self, name, id):
        self.name = name
        self.id = id


class Student(Person):
    def __init__(self, name, id, degree_subject):
        super().__init__(name, id) # this expands the sub-class with super-class features
        self.degree_subject = degree_subject
        
    def greet_student(self): 
        print('Morning {0}, use your id {1} to log into your {2} class'.format(self.name, self.id, self.degree_subject))


class Staff(Person):
    def __init__(self, name, id, position):
        super().__init__(name, id) # this expands the sub-class with super-class features
        self.position = position
        
    def greet_staff(self): 
        print('Morning {0}, use your id {1} to log into your {2} workload'.format(self.name, self.id, self.position))



obj1 = Student('Hermione', 123, 'Charms')
obj1.greet_student()

obj0 = Staff('Snape', 777, 'Defence against the Dark Arts Professor')
obj0.greet_staff()
````

````python
class Animal:
    def __init__(self, role, name, color, hp, atk, defence):
        self.role = role
        self.name = name
        self.color = color
        self.hp = hp   # the animal's health points
        self.atk = atk   # the animal's attack power
        self.defence = defence

    def declare(self):
        print(f"I am {self.role} {self.name}. I am {self.color}. My HP is {self.hp}. \
              My attacking power is {self.atk}. My defence is {self.defence}.")

    def win(self):
        print(f"{self.role} {self.name} Wins!")

    def attack(self,enemy):
        if enemy.hp>self.atk:
            enemy.hp = enemy.hp - self.atk        
        else:
            enemy.hp = 0
# create instance - cat
cat = Animal('Cat', 'Carl', 'blue', 30, 5)  
# create instance - dog
dog = Animal('Dog', 'Doug', 'grey', 28, 7)
````

````python
cat.attack(dog)
print(f"cat hp: {cat.hp},  dog hp: {dog.hp}")
````
